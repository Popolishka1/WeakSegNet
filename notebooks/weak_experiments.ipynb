# Experimentation with weakly-supervised task

In this task we aim to experiment with weakly supervised learning for image segmentation on the Oxford-IIIT Pet Dataset.

The weakly in weakly supervised learning flows from the use of only image-level labels to perform segmentation and not ground truth masks. The image-level labels that we will use are (a) species and breed name, and (b) a tight bounding box (ROI) around the head of the animal, for each image.
